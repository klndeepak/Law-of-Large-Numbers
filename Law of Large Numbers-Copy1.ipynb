```python
# import all libraries required 
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Let us consider throwing a fair dice. The expected value of all possible outcomes is 3.5. 
# As number of trials increase, the avarege tends to true mean. 

results = []
for num_throws in range(1,100000):
    throws = np.random.randint(low=1,high=7, size=num_throws)
    sample_average = throws.mean()
    results.append(sample_average)

df = pd.DataFrame({ 'Throws' : results})
figsize(5, 5)
df.plot(title='Law of Large Numbers for a dice',color='b')
plt.xlabel("Number of throws in sample")
plt.ylabel("Sample Average")
```
